## Display Image

```python
#imread function from cv2 library to read jpeg
jpeg = cv2.imread('D:\\ag.jpg', 1)

#sizing the jpeg
jpeg = cv2.resize(jpeg, (400,580))


jpeg = cv2.cvtColor(jpeg, cv2.COLOR_BGR2RGB)
display(Image.fromarray(jpeg))
```

## Display image using matplotlib (BGR to RGB)

```python
#adjust the plot widtgh and length accordingly
plot.figure(num = 2, figsize=(10.5,8))
#converting from BGR to RGB
plot.imshow(cv2.cvtColor(jpeg, cv2.COLOR_BGR2RGB))

# plt.gca().axes.get_yaxis().set_visible(True)
# plt.gca().axes.get_xaxis().set_visible(True)
plot.show()
```

```python
import cv2
image = cv2.imread('D:\\ag.jpg')
image = cv2.resize(image, (400,580))
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite("D:\\gray.png", gray_image)
displayGray=cv2.imread("D:\\gray.png",1)
display(Image.fromarray(displayGray))
```

```python
inverted_image = 255 - gray_image
cv2.imwrite("D:\\inv.png", inverted_image)
displayInverted=cv2.imread("D:\\inv.png",1)
display(Image.fromarray(displayInverted))
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2.imwrite("D:\\blur.png", blurred)
displayBlur=cv2.imread("D:\\blur.png",1)
display(Image.fromarray(displayBlur))
```

```python
inverted_blurred = 255 - blurred
cv2.imwrite("D:\\invblur.png", inverted_blurred)
displayInvertedBlur=cv2.imread("D:\\invblur.png",1)
display(Image.fromarray(displayInvertedBlur))
```

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imwrite("D:\\Sketch.png", pencil_sketch)
final_sketch=cv2.imread("D:\\Sketch.png",1)
display(Image.fromarray(final_sketch))
```
